```
plt.scatter(data['x'],data['y'],s=250,c=color_list)
```

# Bar chart

bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs) ===>>> อะไรที่ไม่จำเป็นจะมี =

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
```

^ กำหนดขนาด, ความกว้าง ของแต่ละแท่ง

```
plt.bar(x - width/2, men_means, width, label='Men')
plt.bar(x + width/2, women_means, width, label='Women')
```

^ แบบง่ายใช้ plt.bar ได้เหมือนข้างล่าง

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
```

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores') #ตั้งชื่อแกน y
```

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores') #ตั้งชื่อแกน y
ax.set_title('Scores by group and gender') #ตั้งชื่อกราฟ
```

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores') #ตั้งชื่อแกน y
ax.set_title('Scores by group and gender') #ตั้งชื่อกราฟ
ax.set_xticks(x) #ชี้แนวแกน x
ax.set_xticklabels(labels)
ax.legend()
```

```
plt.bar(x - width/2, men_means, width, label='Men')
plt.bar(x + width/2, women_means, width, label='Women')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
```
